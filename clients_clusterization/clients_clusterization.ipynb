```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Mall_Customers.csv')

plt.figure(figsize=(10, 5))
plt.hist(data['Age'], bins=20, rwidth=0.95, color='#264653')
plt.title('Age distribution')
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(data['Annual Income (k$)'], bins=20, rwidth=0.95, color='#264653')
plt.title('Annual Income (k$) distribution')
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(data['Spending Score (1-100)'], bins=20, rwidth=0.95, color='#264653')
plt.title('Spending Score distribution')
plt.show()
```

```python
labelsChart = ['Female', 'Male']

# Female = 0, Male = 1
res = data['Gender'].value_counts()
res = res.rename({0: 'Female', 1: 'Male'})
print("All database:\n" + str(res))

plt.figure(figsize=(12, 8))
plt.pie(data['Gender'].value_counts(), labels = labelsChart, colors=['#2a9d8f', '#e9c46a'], autopct = '%.0f%%')
plt.title('Gender distribution', size=20)
plt.axis('off')
plt.legend()
plt.show()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

le.fit(data['Gender'])
print(le.classes_)
data['Gender'] = le.transform(data['Gender'])

data = data.drop("CustomerID", axis=1)

data.head()
```

```python
import seaborn as sn

corrMatrix = data.corr()
plt.figure(figsize=(10,10))
ax = sn.heatmap(corrMatrix, vmin=-1, vmax=1, center=0, cmap=sn.diverging_palette(20, 220, n=200), 
                square=True, annot=True, fmt='.2f')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right',);
plt.title("Corelation matrix", size=18)
plt.show()

covMatrix = data.cov()
plt.figure(figsize=(10,10))
ax = sn.heatmap(covMatrix, vmin=-100, vmax=600, center=200, cmap=sn.diverging_palette(20, 220, n=200), 
                square=True, annot=True, fmt='.2f')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right');
plt.title("Covariance matrix", size=18)
plt.show()
```

```python
from sklearn.cluster import KMeans

x = data[["Spending Score (1-100)", "Age"]]
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10,8))
sc = plt.scatter(x['Spending Score (1-100)'], x['Age'],c=labels,s=50,cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1],marker='^',c='black',s=200,alpha=0.5)
lp = lambda i: plt.plot([],color=sc.cmap(sc.norm(i)), ms=np.sqrt(len(labels)), mec="none",
                        label="{:g}".format(i), ls="", marker="o")[0]
handles = [lp(i) for i in np.unique(labels)]
plt.xlabel('Spending Score')
plt.ylabel('Age')
plt.title('Spending Score vs Age')
plt.grid(False)
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

def sumVar(x):
    res = 0
    for i in x:
        res += i*100
    return str(res)

pca = PCA(n_components=2)
result = pca.fit_transform(data)
print("Accuracy 2D: " + sumVar(pca.explained_variance_ratio_))
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(result)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(10,8))
sc = plt.scatter(result[:,0],result[:,1],c=labels,s=50,cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1],marker='^',c='black',s=100,alpha=0.5)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid(False)
lp = lambda i: plt.plot([],color=sc.cmap(sc.norm(i)), ms=np.sqrt(len(labels)), mec="none",
                        label="Cluster{:g}".format(i), ls="", marker="o")[0]
handles = [lp(i) for i in np.unique(labels)]
plt.legend()
plt.show()
```

```python
data['cluster'] = labels
cluster0 = data[data["cluster"] == 0]
cluster1 = data[data["cluster"] == 1]
cluster2 = data[data["cluster"] == 2]
cluster3 = data[data["cluster"] == 3]
cluster4 = data[data["cluster"] == 4]

labelsChart = ['Female', 'Male']

# Female = 0, Male = 1
print("Cluster0:\n" + str(cluster0['Gender'].value_counts().rename({0: 'Female', 1: 'Male'})))
print("\nCluster1:\n" + str(cluster1['Gender'].value_counts().rename({0: 'Female', 1: 'Male'})))
print("\nCluster2:\n" + str(cluster2['Gender'].value_counts().rename({0: 'Female', 1: 'Male'})))
print("\nCluster3:\n" + str(cluster3['Gender'].value_counts().rename({0: 'Female', 1: 'Male'})))
print("\nCluster4:\n" + str(cluster4['Gender'].value_counts().rename({0: 'Female', 1: 'Male'})))

plt.figure(figsize=(12, 8))
plt.pie(cluster0['Gender'].value_counts(), labels = ['Male', 'Female'], colors=['#e9c46a', '#2a9d8f',], autopct = '%.0f%%')
plt.title('Gender distribution cluster0', size=20)
plt.axis('off')
plt.legend()
plt.show()

plt.figure(figsize=(12, 8))
plt.pie(cluster1['Gender'].value_counts(), labels = labelsChart, colors=['#2a9d8f', '#e9c46a'], autopct = '%.0f%%')
plt.title('Gender distribution cluster1', size=20)
plt.axis('off')
plt.legend()
plt.show()

plt.figure(figsize=(12, 8))
plt.pie(cluster2['Gender'].value_counts(), labels = labelsChart, colors=['#2a9d8f', '#e9c46a'], autopct = '%.0f%%')
plt.title('Gender distribution cluster2', size=20)
plt.axis('off')
plt.legend()
plt.show()

plt.figure(figsize=(12, 8))
plt.pie(cluster3['Gender'].value_counts(), labels = labelsChart, colors=['#2a9d8f', '#e9c46a'], autopct = '%.0f%%')
plt.title('Gender distribution cluster3', size=20)
plt.axis('off')
plt.legend()
plt.show()

plt.figure(figsize=(12, 8))
plt.pie(cluster4['Gender'].value_counts(), labels = labelsChart, colors=['#2a9d8f', '#e9c46a'], autopct = '%.0f%%')
plt.title('Gender distribution cluster4', size=20)
plt.axis('off')
plt.legend()
plt.show()
```
